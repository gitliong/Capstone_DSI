## Step03-Countvectorizing the Data Collected##

I want to understand what all the regional job descriptions aggregated together would produce as a single 
document itself. 

I have found that items collected in "Paragraphs" typically relate to the work environment or culture of the company. Additionally items that have been collected in "Bullets" relate to the specific technical requirements 
of the job. 

This secion looks to Countvectorize the "Bullets" section specifically. 

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

cvec = CountVectorizer()
```

```python
Para = pd.read_csv('Paragraphs')
```

```python
bullets = pd.read_csv('Bullets')
bullets
```

```python
bullets['0'][258]
```

`cvec.fit()` will be expecting a pandas Series of text objects:

```python
cvec.fit(bullets['0'])
new_corpus = cvec.transform(bullets['0'])
new_corpus
```

```python
new_corpus.todense()
```

```python
df  = pd.DataFrame(new_corpus.todense(),
                   columns=cvec.get_feature_names())

df.head()

df.sort_values('data', ascending=False).head(10)
```

```python
df.sum().sort_values(ascending=False)
```

```python
#clean it up by importing STOP Words

from sklearn.feature_extraction import stop_words
 
print(stop_words.ENGLISH_STOP_WORDS)
```

```python
df.sum().sort_values(ascending=False)
```

```python
#save cleaned up dataframe

df.to_csv('Word Rank')
```
