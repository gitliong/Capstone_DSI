## Proposal ##


The Proposal I will be working on will be my "Indeed.com data scrape to overlay over Google Maps"



## Main Goal of Capstone
**Overarching Goal is to reduce the time it takes to land a job after graduation**


Baseline Goal is 3 months. 


is defined as either "Full Time" or "Contract". 


## Problem Statement

Those looking for a job are at an inherent disadvantage.

This disadvantage starts from the very beginning when applicants are trying to submit their qualifications through an "Applicant Tracking System" and goes on throughout the interview process. The aim of my capstone is to "level the playing field" by providing myself (as a job candidate) the best opportunities to get past the automated "Applicant Tracking System". Additionally once this has been accomplished - I aim to provide myself the best analytical data to go into each interview. 

## Proposed Methods and Models

**How will I get the data**

Web Scraping - To Get Data from Indeed.com

**Models**

Word Count Vectorizing - To be able to compare the data gathered to a resume

Tfid

Logistic Regression - To see how close of a match

Feature Engineering

Google Maps
