**Using the Slugs to visit each Link and get the full Job Description**

In the section below we scrape Indeed.com. 

However we append each slug to the end of Indeed.com and that will take us to each web page with 
the full job description. 

As always - during this specific process - we create a blank list. 

```python
extended_descriptions = []

list_of_dicts = []
```

```python
for r in slugs[:30]:
    new_url = 'http://www.indeed.com' + r
    print('Requesting content from ' + new_url) # you can add + '...' 
    res = requests.get(new_url)
    # print('Converting content from the res object.')
    soup = BeautifulSoup(res.content, 'lxml')
    extended_descriptions.append(soup)
    sleep(3)
    print('Appending soup...')
```

```python
list_of_dicts
```

```python
len(extended_descriptions)
```

```python
for each in extended_descriptions:
    print(type(each))
```

**Scraping specific sections of the 'slug' page**

In the sections below I will be scraping all the text that have been bolded, 'b', all the text that are 
paragraphs, 'p' and all the text that have been bulleted, 'li'.

```python
for page in extended_descriptions:
    for each in page.find_all('b'):
        print(each.get_text())
```

```python
list_of_para = []
for page in extended_descriptions:
    for each in page.find_all('p'):
        list_of_para.append(each.get_text())
```

```python
list_of_li = []

for page in extended_descriptions:
    for each in page.find_all('li'):
        list_of_li.append(each.get_text())
```

**Transferring the scraped Paragraphs from the Job Description to a CSV**

In this section I will transfer the paragraphs collected to a CSV.

I will then read the CSV to perform some cleanup on the data so that the final saved CSV is clean data ready to use. 

```python
pd.DataFrame(list_of_para).to_csv('Paragraphs', index=False)
```
