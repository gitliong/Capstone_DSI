```python
indeed_cities = ['Atlanta']
len(indeed_cities)
```

**Creating a List**

The important thing to remember here is to initially create a blank list. 
The list is then appended with all the results from the scrape. 
Note to put a sleep function so that we are not kicked out.

```python
max_results_per_city = 100

results = []

for city in indeed_cities:
    for start in range(0, max_results_per_city, 100):
        url = "https://www.indeed.com/jobs?as_and=data+scientist+python&as_phr=&as_any=&as_not=&as_ttl=&as_cmp=&jt=all&st=\
               &salary=&radius=25&l= + Atlanta + &fromage=any&limit=100&start=" + str(start) + "&sort=&psf=advsrch"
        html = requests.get(url)
        soup = BeautifulSoup(html.text, 'html.parser')
        for result in soup.find_all('div', {'class':' row result'}):
            results.append(result)
        sleep(1)
```

```python
results
```

**Creating Functions for specific scrapes**

```python
def get_age(result):
    try:
        return result.find('span', {'class':'date'}).text
    except:
        return 'NA'
```

```python
def get_loc(result):
    try:
        return result.find('span', {'class':'location'}).text
    except:
        return 'NA'
```

```python
def get_comp(result):
    try:
        return result.find('span', {'class':'company'}).text
    except:
        return 'NA'
```

```python
def get_job(result):
    try:
        return result.find('a', {'data-tn-element':'jobTitle'}).text
    except:
        return 'NA'
```

```python
def get_sal(result):
    try:
        return result.find('td', {'class':'snip'}).find('nobr').text
    except:
        return 'NA'
```

```python
def get_desc(result):
    try:
        return result.find('span', {'class':'summary'}).text
    except:
        return 'NA'
```

```python
df = pd.DataFrame(columns=['location', 'title', 'company', 'salary', 'summary', 'age'])
```

```python
for entry in results:
    age = get_age(entry)
    location = get_loc(entry)
    title = get_job(entry)
    company = get_comp(entry)
    salary = get_sal(entry)
    desc = get_desc(entry)
    df.loc[len(df)] = [location, title, company, salary, desc, age]
```

**Capturing the Slugs**

Slugs are the tail end of an internet address that tells a link what specific web page to go to. 

The process below will be to identify that specific tail end which is 'a' and 'href' and then to append them 
to a list. Again note the important step of creating a blank list. 


```python
slugs = []

for each in results:
    slugs.append(each.find('a')['href'])
    
len(slugs)
```

```python
slugs
```

```python
df['slugs'] = slugs
```
